# 리스트를 스캔하는 4가지 방법

```python
# 2C_1
# 리스트의 모든 원소를 스캔하기(원소 수를 미리 파악)

x = ['John', 'George', 'Paul', 'Ringo']

for i in range(len(x)):
    print(f'x[{i}] = {x[i]}')
```

```python
# 2C-2
# 리스트의 모든 원소를 enumerate() 함수로 스캔하기

x = ['John', 'George', 'Paul', 'Ringo']

for i, name in enumerate(x):
    print(f'x[{i}] = {name}')
```

```python
# 2C-3
# 리스트의 모든 원소를 enumerate() 함수로 스캔하기(1부터 카운트)

x = ['John', 'George', 'Paul', 'Ringo']

for i, name in enumerate(x, 1): # 시작하는 번호를 부여한다는 것정도만 주목하면 될듯
    print(f'x[{i}] = {name}')
```

```python
# 2C-4
# 리스트의 모든 원소를 스캔하기(인덱스 값을 사용하지 않음)

x = ['John', 'George', 'Paul', 'Ringo']

for name in x:
    print(name)
```

# 리스트를 역순으로 정렬하기

```python
# 뮤터블 시퀀스 원소를 역순으로 정렬 (뮤터블이여서 원소를 조작할 수 있기 때문에 조건에 추가함)

from typing import Any, MutableSequence # annotation에 사용하기 위해서 가져옴

def reverse_array(a: MutableSequence) -> None:
    
    n = len(a)
    for i in range(n//2):
        a[i], a[n-1-i] = a[n-1-i], a[i]
```

```python
if __name__ == '__main__':
    print('배열 원소를 역순으로 정렬합니다.')
    nx = int(input('원소 수를 입락하세요: '))
    x = [None] * nx

    for i in range(nx):
        x[i] = input(f'x[{i}]의 원소값을 입력하세요: ')

    reverse_array(x)

    print('리스트를 역순으로 정렬했습니다')
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
```

```python
# 내장 함수를 사용하는 방법(자신을 역순으로 바꾸는 방법, inplace)

x = [1,2,3,4, 5]
x.reverse()
print(x)
```

```python
# 내장 함수를 사용하는 방법2 (자신은 나누고, 역순으로 정렬된 객체를 생성)

print(x)
x_reversed = list(reversed(x)) # 정확히는 iterator를 반환하는것으로, interable한 list 객체를 만들어주기 위해서는 list로 타입 변환이 필요하다
print(x_reversed)
```

# 기수 변환하기(n진수 구하기)

```python
# 10진수 정숫값을 입력받아 2~36진수로 변환하여 출력하기

def card_conv(x: int, r: int) -> str:
    """ 정수값 x를 r진수로 변환한 뒤 그 수를 나타내는 문자열 반환 """
    
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ' # 36진법까지 사용할 수 있는 기수 모음
    
    while x > 0:
        d += dchar[x % r] # 나머지에 해당되는 수를 꺼내는 역할
        x //= r           # 몫(묶음)개수
        
    return d[::-1]        # 전부 역순으로 출력
```

```python
if __name__ == '__main__': # 지금 파일이 내가 작업하고 있는 파일이라면,
    print('10진수를 n진수로 변환합니다.')
    
    while True:
        
        while True:
            no = int(input('변환할 값으로 음이 아닌 정수를 입력하세요: '))
            if no > 0:
                break
            
        while True:
            cd = int(input('어떤 진수로 변환할까요?: ')) 
            if 2 <= cd <= 36:
                break
                
        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')
        
        retry = input('한번 더 변환할까요? (Y/N): ')
        if retry in {'N', 'n'}:
            break
```

### __name__ 보충설명
- \_\_name\_\_ 속성은 파일의 이름이 저장되는 변수
- 만약, 어떤 파일을 import 를 통해서 불러오면 불러온 파일의 \_\_name\_\_에는 해당 파일의 이름이 적혀있음
- 주의할 점은 현재 파일의 이름이 hello라고 하면 \_\_name\_\_ 에는 __\_\_main\_\___ 값이 들어가게 됨
- \_\_main\_\_ 이라는 값은 현재 실행하는 파일을 의미함.

# 소수 나열하기

```python
# 1000 이하의 소수 나열하기

counter = 0 # 나눗셈 횟수 저장

for n in range(2, 1000+1):
    for i in range(2,n):
        counter += 1
        if n%i==0:
            """소수가 아닌 경우"""
            break
    else:        
        """끝까지 나누어지지 않으면 출력"""
        print(f'{n}')
    
print(f'나눗셈을 실행한 횟수: {counter}')
```

```python
# 소수란 결국 더 작은 소수들로 나누어지지 않는 존재들...
# 알고리즘 개선1

counter = 0 # 나눗셈 횟수 저장
prime = [2]

for n in range(3, 1000+1, 2): # 짝수는 2(소수)의 배수니까 제외
    for i in prime[1:]:
        counter += 1
        if n%i==0:
            """소수가 아닌 경우"""
            break
    else:        
        """끝까지 나누어지지 않으면 출력"""
        prime.append(n)
        print(f'{n}')
    
print(f'나눗셈을 실행한 횟수: {counter}')
```

```python
# 약수로 나누어떨어지는 것을 이용해서 소수란, 제곱근아래의 수로 떨어지지 않는 수
# 알고리즘 개선 2

"""알고리즘 1은 n보다 작은 소수 전부를 활용했다면, 알고리즘 2는 제곱근보다 작은 소수들만을 활용"""

counter = 0 # 연산 횟수 저장
prime = [2]

for n in range(3, 1000+1, 2): # 짝수는 2(소수)의 배수니까 제외
    for i in prime[1:]:
        if n >= i*i:           # i가 n의 제곱근이 작은 경우에
            counter += 2       # 제곱, 나누기 연산이 각각 한번씩이므로 2번 카운트
            if n%i ==0:
                """소수가 아닌 경우"""
                break
    else:        
        """끝까지 나누어지지 않으면 출력"""
        prime.append(n)
        print(f'{n}')
    
print(f'나눗셈을 실행한 횟수: {counter}')
```
